```python
df = df[['Date','Confirmed','Tested','First Dose Administered']]
df
```

```python
df1= df.copy()

df1["Daily_confirmed"] = ""
df1["Daily_Tested"] = ""
df1["Average_of_7days_Confirmed"] = ""
df1["Average_of_7days_Tested"] = ""
df1["Daily_first_dose"] = ""
# df1["CIR"] = ""

for i, rowi in df1.iterrows():
    if(i>0):
        df1["Daily_confirmed"][i]= df1["Confirmed"][i] -df1["Confirmed"][i-1]
        df1["Daily_Tested"][i]= df1["Tested"][i] -df1["Tested"][i-1]
        df1["Daily_first_dose"][i]= df1["First Dose Administered"][i]-df1["First Dose Administered"][i-1]
    else:
        df1["Daily_confirmed"][i]= df1["Confirmed"][i]
        df1["Daily_Tested"][i]=   df1["Tested"][i]
        df1["Daily_first_dose"][i]=   df1["First Dose Administered"][i]
        
df1
```

```python
for i, rowi in df1.iterrows():
    count_confirmed=0
    c1=0
    count_Tested=0
    c2=0
    for j in range(i-6,i+1):
        if(j<0):
            c1=0
            c2=0
        else:
            c1=df1["Daily_confirmed"][j]
            c2=df1["Daily_Tested"][j]
        
        count_confirmed=count_confirmed+c1
        count_Tested=count_Tested+c2

    df1["Average_of_7days_Confirmed"][i]=count_confirmed/7
    df1["Average_of_7days_Tested"][i]=count_Tested/7
    
```

```python
df1= df1[(df1["Date"]>='2021-03-16') & (df["Date"]<='2021-04-26')]
df1
```

```python
eps=0.66
alpha= 1/5.8
gamma= 1/5
def equations(S_t,E_t,I_t,R_t,B_t,V_t,W_t):
    global N
    global eps
    global alpha
    global gamma 
    S_t_1 = S_t - ((B_t * S_t * I_t)/N) - (eps * V_t) + W_t
    
    E_t_1 = E_t + ((B_t * S_t * I_t)/N) -(alpha * E_t)
    
    I_t_1 = I_t + (alpha * E_t) - (gamma * I_t)
    
    R_t_1 = R_t + (gamma * I_t) + (eps * V_t) - W_t
    
    return S_t_1, E_t_1, I_t_1, R_t_1 
    
```

```python
N = 70000000
S_t= 0.7 *N                                #60900000 
E_t=.12 *N   #.134 *N                                        #7000000
I_t= .02 *N     #0.01 *N   #700000 *2
R_O = .16 *N
CIR_O= 13
B_t = 2
 
# S_t = 10000000
#  E_t = 10000000
#  I_t = 12* 962339         
#  R_t = 700000
                            
R_t=R_O
T_O=df1["Average_of_7days_Tested"].iloc[0]
# print(T_O)

S_t_list=[]
E_t_list=[]
I_t_list=[]
R_t_list=[]
CIR_t_list=[]
e_t_list=[]
c_t_list=[]
for i, rowi in df1.iterrows():
    val=rowi["Average_of_7days_Tested"]
    c_t_list.append(rowi["Average_of_7days_Confirmed"])
    
    CIR_t= CIR_O *(T_O/val)
    
    CIR_t_list.append(CIR_t)
    
    if(rowi["Date"]>='2021-03-16' and rowi["Date"]<'2021-04-15'):
        W_t= R_O/30
    elif(rowi["Date"]>='2021-04-15' and rowi["Date"]<='2021-04-26'):
        W_t=0
        
    V_t = rowi['Daily_first_dose']
#     print(V_t)
    
#     print(S_t)
    
    S_t_1, E_t_1, I_t_1, R_t_1 = equations(S_t, E_t, I_t, R_t, B_t,V_t,W_t)
#     print(S_t_1 - S_t, S_t_1, S_t)
    
    S_t_list.append(S_t_1)
    E_t_list.append(E_t_1)
    I_t_list.append(I_t_1)
    R_t_list.append(R_t_1)
    
#     e_t =alpha * (E_t_1/CIR_t)
    e_t =(E_t_1/CIR_t)
    e_t_list.append(e_t)

    S_t, E_t, I_t, R_t  = S_t_1, E_t_1, I_t_1, R_t_1
```

```python
data_1 = {'S_t':S_t_list,
        'E_t':E_t_list,
       'I_t':I_t_list,
         'R_t':R_t_list,
          'CIR_t': CIR_t_list,
          'e_t':e_t_list,
          'c_t':c_t_list
         }
 
# Create DataFrame
df4 = pd.DataFrame(data_1)
 
#Print the output.
e_t_average= df4.e_t.rolling(7,min_periods=1).mean()
df4['e_t_average'] = e_t_average
#Multiplying rolling error with alpha

df4["e_t_average"] = alpha * df4["e_t_average"]
df4
```
