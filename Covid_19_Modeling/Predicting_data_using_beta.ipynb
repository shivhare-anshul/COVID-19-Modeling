```python
import pandas as pd
pd.options.mode.chained_assignment = None
import numpy as np
```

```python
eps=0.66
alpha= 1/5.8
gamma= 1/5
def equations(S_t,E_t,I_t,R_t,B_t,V_t,W_t):
    global N
    global eps
    global alpha
    global gamma 
    S_t_1 = S_t - ((B_t * S_t * I_t)/N) - (eps * V_t) + W_t
    
    E_t_1 = E_t + ((B_t * S_t * I_t)/N) -(alpha * E_t)
    
    I_t_1 = I_t + (alpha * E_t) - (gamma * I_t)
    
    R_t_1 = R_t + (gamma * I_t) + (eps * V_t) - W_t
    
    return S_t_1, E_t_1, I_t_1, R_t_1 
```

```python
df = pd.read_csv('COVID19_data.csv')
df = df[['Date','Confirmed','Tested','First Dose Administered']]

# Making copy of datafarme
df1= df.copy()

# Adding few empty columns in dataframe
df1["Daily_confirmed"] = ""
df1["Daily_Tested"] = ""
df1["Average_of_7days_Confirmed"] = ""
df1["Average_of_7days_Tested"] = ""
df1["Daily_first_dose"] = ""

# **********************************************************************************************************
# Counting daily confirmed and daily tested cases
for i, rowi in df1.iterrows():
    if(i>0):
        df1["Daily_confirmed"][i]= df1["Confirmed"][i] -df1["Confirmed"][i-1]
        df1["Daily_Tested"][i]= df1["Tested"][i] -df1["Tested"][i-1]
        df1["Daily_first_dose"][i]= df1["First Dose Administered"][i]-df1["First Dose Administered"][i-1]
    else:
        df1["Daily_confirmed"][i]= df1["Confirmed"][i]
        df1["Daily_Tested"][i]=   df1["Tested"][i]
        df1["Daily_first_dose"][i]=   df1["First Dose Administered"][i]
        
# ***********************************************************************************************************
# Running average of last 7 days, for both Daily confirmed and Daily tested.
        
for i, rowi in df1.iterrows():
    count_confirmed=0
    c1=0
    count_Tested=0
    c2=0
    for j in range(i-6,i+1):
        if(j<0):
            c1=0
            c2=0
        else:
            c1=df1["Daily_confirmed"][j]
            c2=df1["Daily_Tested"][j]
        
        count_confirmed=count_confirmed+c1
        count_Tested=count_Tested+c2

    df1["Average_of_7days_Confirmed"][i]=count_confirmed/7
    df1["Average_of_7days_Tested"][i]=count_Tested/7
            
# *************************************************************************************************************
df1= df1[(df1["Date"]>='2021-03-16') & (df1["Date"]<='2021-09-11')]
df1 = df1.reset_index(drop=True)
#df1
```

```python
#Predicting data from march 16th till 20TH september using optimized parameters
def Generating_predictions(B_t,S_t,E_t,I_t,R_O,CIR_O):
    R_t=R_O
    
    S_t_list=[]
    E_t_list=[]
    I_t_list=[]
    R_t_list=[]
    W_t_list=[]
    V_t_list=[]

    for i in range(0,189):

        if(i<=30):
#             print(i)
            W_t= R_O/30
            V_t = df1["Daily_first_dose"][i]
            W_t_list.append(W_t)
            V_t_list.append(V_t)
        elif(i>30 and i<=179):
            W_t=0
            V_t = df1["Daily_first_dose"][i]
            W_t_list.append(W_t)
            V_t_list.append(V_t)
        else:
#             print(len(R_t_list)-179)
            W_t=R_t_list[len(R_t_list)-180] + eps * V_t_list[len(V_t_list)-180]
            
            V_t=200000                             #seeting vaccinatio per day as 2 lakhs after 
            V_t_list.append(V_t)
            W_t_list.append(W_t)



        S_t_1, E_t_1, I_t_1, R_t_1 =equations(S_t,E_t,I_t,R_t,B_t,V_t,W_t)

        S_t_list.append(S_t_1)
        E_t_list.append(E_t_1)
        I_t_list.append(I_t_1)
        R_t_list.append(R_t_1)
        
        S_t, E_t, I_t, R_t= S_t_1, E_t_1, I_t_1, R_t_1


    return S_t_list,E_t_list,I_t_list,R_t_list,W_t_list,V_t_list
    
```

```python
# prediction using beta1
B_t=B
S_t_list,E_t_list,I_t_list,R_t_list,W_t_list,V_t_list= Generating_predictions(B_t,S_t,E_t,I_t,R_O,CIR_O)
data_1 = {'S_t':S_t_list,
                'E_t':E_t_list,
               'I_t':I_t_list,
                 'R_t':R_t_list,

                 }

        # Create DataFrame
df_beta1 = pd.DataFrame(data_1)
# print(S_t_list)
df_beta1
```
